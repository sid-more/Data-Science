# Setting lable with Object Oriented Technique

```python
fig = plt.figure(figsize = [5,5])
axes = fig.add_axes([0,0,1,1])
axes.set_xlabel('x-axis')
axes.set_ylabel('y-axis')
axes.set_title("plot")
axes.plot(x,y)
```

# Styling Plot

```python
fig = plt.figure(figsize = [5,5])
axes = fig.add_axes([0,0,1,1])
axes.set_xlabel('x-axis')
axes.set_ylabel('y-axis')
axes.set_title("plot")
axes.plot(x,y, color="g" ,lw='10', alpha = 0.5, linestyle='--' )
```

# Markups

```python
fig = plt.figure(figsize = [5,5])
axes = fig.add_axes([0,0,1,1])
axes.set_xlabel('x-axis')
axes.set_ylabel('y-axis')
axes.set_title("plot")
axes.plot(x,y , marker='o' , markersize=10 , markerfacecolor='g' ,markeredgecolor='y',markeredgewidth='5')
```

# limits

```python
fig = plt.figure(figsize = [5,5])
axes = fig.add_axes([0,0,1,1])
axes.set_xlim([0,10])
axes.set_ylim([0,60])
axes.set_title("plot")
axes.plot(x,y)
```
