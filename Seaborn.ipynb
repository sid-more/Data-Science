# Two Parameters

# Factorplot ,Swarmplot,stripplot

```python
sns.factorplot(x='smoker',y='tip',data=tips,kind="point")
#we can change kind to box and bar and more
```

```python
sns.swarmplot(x='smoker',y='tip',data=tips,hue ='sex')
```

```python
sns.swarmplot(x='smoker',y='tip',data=tips,hue ='sex',palette='cool')
#palette:- magma ,coolwarm,
```

```python
sns.stripplot(x='smoker',y='tip',data=tips,hue ='sex',palette='coolwarm')
#palette:- magma ,coolwarm,
```
